reshape 操作:高维降到1维，再从最低维度开始组合构建。  
深度学习中，每个维度都有“含义"的，简单低维到高维的拼接是不行的。

```python
from torchvision.datasets import CIFAR10
from torchvision.transforms import ToTensor
import matplotlib.pyplot as plt
```

```python
test = CIFAR10(r"F:\VsConde-Python\chen\data", train=False, download=False, transform=ToTensor())
img, lbl = test[123]
```
